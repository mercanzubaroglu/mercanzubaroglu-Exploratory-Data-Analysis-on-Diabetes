```python
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

```python
#Veri Setini Yükleme
df = pd.read_csv("C:/Users/aziz-/Downloads/archive/diabetes.csv")

```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
# 0 olan değerleri kontrol et (eksik sayılabilecekler)
cols_with_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols_with_zero].eq(0).sum()
```

```python
sns.countplot(x='Outcome', data=df, palette='Set2')
plt.title("Diyabet Durumu (0 = Yok, 1 = Var)")
plt.show()

# Yüzdelik dağılım
df['Outcome'].value_counts(normalize=True) * 100
```

```python
sns.histplot(data=df, x='Age', hue='Outcome', kde=True, bins=30)
plt.title("Yaş Dağılımı ve Diyabet İlişkisi")
plt.show()

sns.histplot(data=df, x='Glucose', hue='Outcome', kde=True, bins=30)
plt.title("Glukoz Seviyesi ve Diyabet Riski")
plt.show()
```

```python
cols = ['Glucose', 'BloodPressure', 'BMI', 'Insulin', 'Age']

for col in cols:
    sns.boxplot(x='Outcome', y=col, data=df, palette='pastel')
    plt.title(f"{col} - Diyabet İlişkisi")
    plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu', linewidths=0.5)
plt.title("Korelasyon Matrisi")
plt.show()
```
